```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

def draw_plot():
    # Read data from file
    df = pd.read_csv('epa-sea-level.csv')

    # Create scatter plot
    # Extracting columns for the scatter plot
    year = df['Year']
    sea_level = df['CSIRO Adjusted Sea Level']

    # Creating the scatter plot
    plt.figure(figsize=(20, 10))
    plt.scatter(year, sea_level, color='blue', alpha=0.7)

    # Adding labels and title
    plt.xlabel('Year')
    plt.ylabel('CSIRO Adjusted Sea Level (Inches)')
    plt.title('CSIRO Adjusted Sea Level over Time')

    # Displaying the plot
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Create first line of best fit
    # Extracting columns for the scatter plot
    year = df['Year']
    sea_level = df['CSIRO Adjusted Sea Level']

    # Creating the scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(year, sea_level, color='blue', alpha=0.7, label='Data')

    # Calculating the line of best fit
    slope, intercept, r_value, p_value, std_err = linregress(year, sea_level)
    # Predicting sea level in 2050
    sea_level_2050 = round(slope * 2050 + intercept,5)

    # Plotting the line of best fit
    plt.plot(year, slope * year + intercept, color='red', label='Line of Best Fit')

    # Plotting the prediction for 2050
    plt.scatter(2050, sea_level_2050, color='green', label='Predicted Sea Level in 2050')
    
    # Adding labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches))')
    plt.title('Rise in Sea Level')

    # Adding legend
    plt.legend()

    # Displaying the plot
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print("Predicted sea level in 2050:", sea_level_2050, "Inches")


    # Create second line of best fit
    # Extracting data from year 2000 to the most recent year
    df_recent = df[df['Year'] >= 2000]

    # Extracting columns for the scatter plot
    year_recent = df_recent['Year']
    sea_level_recent = df_recent['CSIRO Adjusted Sea Level']

    # Creating the scatter plot
    plt.figure(figsize=(20, 10))
    plt.scatter(year_recent, sea_level_recent, color='blue', alpha=0.7, label='Data')

    # Calculating the line of best fit for recent data (since 2000)
    slope_recent, intercept_recent, _, _, _ = linregress(year_recent, sea_level_recent)

    # Predicting sea level in 2050
    sea_level_2050_recent = round(slope_recent * 2050 + intercept_recent,5)

    # Plotting the line of best fit for recent data
    plt.plot(year_recent, slope_recent * year_recent + intercept_recent, color='red', label='Line of Best Fit (2000-Present)')

    # Plotting the prediction for 2050 using recent data
    plt.scatter(2050, sea_level_2050_recent, color='green', label='Predicted Sea Level in 2050 (2000-Present)')

    # Adding labels and title
    plt.xlabel('Year')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')

    # Adding legend
    plt.legend()

    # Displaying the plot
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    print("Predicted sea level in 2050 (2000-Present):", sea_level_2050_recent, "Inches")
    
    # Save plot and return data for testing (DO NOT MODIFY)
    plt.savefig('sea_level_plot.png')
    return plt.gca()
```
